## Le nombre théorique de personnes qui pourraient être nourries

```python
besoins_kcal_j = 2500 # besoins journaliers moyen en kilocalories
besoins_kcal_an= 365* besoins_kcal_j
```

```python
dispo_ali2017 = pd.merge(population.loc[population['Année'] == 2017,["Zone", "Valeur"]],
                               dispo_ali, on='Zone')
dispo_ali2017.head()
```

```python
dispo_ali2017['Disponibilité alimentaire (Kcal/an)'] = dispo_ali2017['Disponibilité alimentaire (Kcal/personne/jour)'] \
* 365 * dispo_ali2017['Valeur']
```

```python
population_mondial2017 = population.loc[population['Année'] == 2017,"Valeur"].sum()

population_mondial2017
```

```python
proportion_pop = round((dispo_ali2017['Disponibilité alimentaire (Kcal/an)'].sum() / besoins_kcal_an) / population_mondial2017*100, 2)
proportion_pop
```

```python
dispo_ali2017['Disponibilité alimentaire (Kcal/an)'].sum()

print("Si on souhaite savoir combien de personne on pourrait nourrir par an, il faut diviser la disponibilité alimentaire kcal mondial sur un an par le besoin kcal mondial par an. Voici le resultat pour l'année 2017 :", "{:.0f}".format(dispo_ali2017['Disponibilité alimentaire (Kcal/an)'].sum() / besoins_kcal_an) ,"ce qui correspond à", proportion_pop, "% de la population mondial")
```

## Le nombre théorique de personnes qui pourraient être nourries grâce aux vegetaux

```python
#ici nous faisons une jointure entre population et dispo_ali en ne selectionnant que les données avec l'origine "vegetale"
kcal_dispo_vegetaux = pd.merge(population.loc[population['Année'] == 2017,["Zone", "Valeur"]],
                               dispo_ali.loc[dispo_ali['Origine'] == "vegetale"], on='Zone')
kcal_dispo_vegetaux.head()
```

```python
kcal_dispo_vegetaux['Disponibilité alimentaire (Kcal/an)'] = kcal_dispo_vegetaux['Disponibilité alimentaire (Kcal/personne/jour)'] \
* 365 * kcal_dispo_vegetaux['Valeur']
kcal_dispo_vegetaux.head()

#nous avons à présent une nouvelle colonne Disponibilité alimentaire '(kcal/an)' pour faciliter le calcul .sum()
```

```python
print("Si on souhaite savoir combien de personne on pourrait nourrir par an, il faut diviser la disponibilité alimentaire kcal mondial des vegetaux sur un an par le besoin kcal mondial sur un an. Voici le resultat pour l'année 2017 :", "{:.0f}".format(kcal_dispo_vegetaux['Disponibilité alimentaire (Kcal/an)'].sum() / besoins_kcal_an))
```

## Disponibilité intérieure

```python

#Production + Importations – Exportations + Variation de stock = Disponibilité intérieure = Semences + Pertes + Nourriture + Aliments pour animaux + Traitement + Autres utilisations.


alim_animal = (dispo_ali['Aliments pour animaux'].sum() / dispo_ali['Disponibilité intérieure'].sum())*100

print("La proportion pour l'alimentation animal sur l'année 2017 est de", "{:.2f}".format(alim_animal),"%")
```

```python
alim_perdu = (dispo_ali['Pertes'].sum() / dispo_ali['Disponibilité intérieure'].sum())*100
print("La proportion pour l'alimentation perdu sur l'année 2017 est de", "{:.2f}".format(alim_perdu),"%")
```

```python
alim_hu = (dispo_ali['Nourriture'].sum() / dispo_ali['Disponibilité intérieure'].sum())*100
print("La proportion pour l'alimentation humaine sur l'année 2017 est de", "{:.2f}".format(alim_hu),"%")
```

## Répartition des céréales dans l'alimentation humaine et animale

```python
cereales=['Blé et produits', 'Riz et produits', 'Orge et produits',
       'Maïs et produits', 'Seigle et produits', 'Avoine',
       'Millet et produits', 'Sorgho et produits', 'Céréales, Autres']
```

```python
liste_cereales = dispo_ali2017[dispo_ali2017['Produit'].isin(cereales)]
```

```python
liste_cereales.head()
```

```python
ali_cereale_animaux = liste_cereales['Aliments pour animaux'].sum() / liste_cereales['Disponibilité intérieure'].sum()*100
print("Ici nous pouvons constater que","{:.2f}".format(ali_cereale_animaux),"% des céréales étaient désintés à l`alimentation animal")
```

```python
ali_cereale_homme = liste_cereales['Nourriture'].sum() / liste_cereales['Disponibilité intérieure'].sum()*100
print("Ici nous pouvons constater que","{:.2f}".format(ali_cereale_homme),"% des céréales étaient désintés à l`alimentation humaine")
```
